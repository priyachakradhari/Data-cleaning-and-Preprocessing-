# About Dataset:

The Netflix Movies and TV Shows dataset contains information about the content available on Netflix as of the latest update. It includes details such as the title, type (Movie or TV Show), director, cast, country of origin, release year, date added to Netflix, content rating, duration, genres, and a brief description.
This dataset is commonly used for data cleaning, exploratory data analysis (EDA), and recommendation system projects. It helps analyze trends in content, popular genres, and Netflix's global content strategy.

 Dataset Overview:

 show_id        : Unique identifier for each title.

 type           : Movie or TV Show.

 title          : Title of the content.

 director       : Director's name (many missing).

 cast           : Main cast members (text list, partially missing).

 country        : Country of production (some missing, multiple possible).

 date_added     : Date added to Netflix (needs datetime conversion).

 release_year   : Original release year of the title.

 rating         : Age rating (e.g., PG, TV-MA).

 duration       : Movie length (minutes) or number of seasons.

 listed_in      : Genre/categories (comma-separated).

 description    : Short summary of the title.

## Handling Numerical Data in Pandas (Theory)

In data preprocessing, numerical columns often require special attention to ensure the dataset is clean and suitable for analysis or modeling. Even though the Netflix dataset does not contain significant numerical features, the following are standard steps used when working with numerical data in Pandas:

Identifying Missing Values:
Use .isnull() and .sum() to detect missing values in numerical columns.

Handling Missing Values:

Mean: Used for normally distributed data (no extreme outliers).

Median: Preferred for skewed data or when outliers are present.

Mode: Useful when dealing with categorical-type numbers or very few unique values.

Example:

python

df['column'].fillna(df['column'].mean(), inplace=True)

---
## Standardization (Feature Scaling):

This process transforms the data to have a mean of 0 and standard deviation of 1. It is essential when numerical features vary in scale, especially before applying machine learning models.

Example using StandardScaler from scikit-learn:

In python,

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df['column_scaled'] = scaler.fit_transform(df[['column']])

These steps help ensure that numerical data is clean, consistent, and ready for further analysis or modeling.

# Important Question

### 1. What are missing values and how do you handle them?

* Missing values = empty/null data in fields (e.g., `director`, `cast`, `country`).
* Handled using:

  * `'Unknown'` or `'Not Available'` for text fields.
  * `mode()` for low-missing columns like `rating`, `date_added`.

---

### 2. How do you treat duplicate records?

* Used `df.drop_duplicates()` to remove exact duplicate rows.
* Ensures data integrity and prevents repeated entries in analysis.

---

### 3. Difference between `dropna()` and `fillna()` in Pandas?

* `dropna()` removes rows/columns with missing values.
* `fillna()` replaces missing values with a specific value (mean, median, mode, etc.).

---

### 4. What is outlier treatment and why is it important?

* Outliers = values that are abnormally high/low compared to the rest.
* Important to treat to avoid skewed results.
* Not needed in Netflix dataset (no numerical runtime column analyzed).

---

### 5. Explain the process of standardizing data.

* Standardization = scaling data to mean = 0 and std = 1.
* Done using `StandardScaler` in Pandas/Sklearn.
* Not applicable to this dataset due to lack of numerical columns.

---

### 6. How do you handle inconsistent data formats (e.g., date/time)?

* Used `pd.to_datetime(df['date_added'], errors='coerce')`.
* Ensures consistent and usable datetime format for analysis.

---

### 7. What are common data cleaning challenges?

* Handling missing or duplicate data.
* Inconsistent text/categorical formatting.
* Wrong data types (e.g., object instead of datetime).
* Choosing the right strategy for filling values.

---

### 8. How can you check data quality?

* Use `.info()`, `.isnull().sum()`, `.duplicated().sum()`, and `.dtypes` in Pandas.
* Visually inspect unique values and distributions.
